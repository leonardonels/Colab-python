2. Crea una nuova colonna chiamata **Livello Esperienza**, utilizzando la colonna **Anni di esperienza**. Se l'esperienza è minore di 5 anni, mappa il valore a "Junior", tra 5 e 10 a "Mid" e più di 10 a "Senior".

```python
def find_level(anni):
  if anni < 5:
    return 'Junior'
  elif anni <= 10:
    return 'Mid'
  else:
    return 'Senior'
```

```python
df['Livello Esperienza'] = df['Anni di esperienza'].map(find_level)
```

```python
df
```

```python
df['Livello Esperienza_cut'] = pd.cut(df['Anni di esperienza'].values, [0,5,11,100], right=False, labels = ['Junior', 'Mid', 'Senior'])
```

3. Sostituisci i nomi dei dipartimenti con le seguenti abbreviazioni: Vendite -> 'VEN', Information Technology -> 'IT', Human Resources -> 'HR'. Quanti dipendenti ci sono per ogni dipartimento?

```python
dip = {'Vendite': 'VEN', 'Information Technology': 'IT', 'Human Resources': 'HR'}
```

```python
df['Dipartimento'] = df['Dipartimento'].map(dip)
```

```python
df['Dipartimento'].value_counts()
```

4. Crea un'altra colonna con i valori di età discretizzati in 5 gruppi, a quale gruppo di età appartiene la maggior parte dei dipendenti?

```python
df['Età_discr'] = pd.cut(df['Età'].values, 5)#.codes
```

```python
df['Età_discr'].value_counts().idxmax()
```

5. Sostituisci i valori mancanti della colonna età con il valore medio di età per i dipendenti dello stesso settore.

```python
df[df['Età'].isnull()]
```

```python
mean_ven = df[df['Dipartimento'] == 'VEN'].dropna()['Età'].mean()
mean_hr = df[df['Dipartimento'] == 'HR'].dropna()['Età'].mean()
```

```python
df.iloc[0,2] = mean_ven
df.iloc[5,2] = mean_hr
```

```python
df
```

```python
mean_age_by_department = df.groupby('Dipartimento')['Età'].mean()

def fill_missing_age(row):
  if pd.isnull(row['Età']):
    return mean_age_by_department.get(row['Dipartimento'])
  else:
    return row['Età']

df['Età'] = df.apply(fill_missing_age, axis=1)
```

6. Verifica se, in media, lo stipendio di chi vive a Milano è superiore rispetto a quello di chi vive a Roma.

```python
mean_milano = df[df['Città']=='Milano']['Stipendio'].mean()
```
