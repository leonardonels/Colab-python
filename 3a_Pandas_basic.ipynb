Other options:
- ***loc***: Access a group of rows and columns by label(s).
- ***iloc***: Purely integer-location based rows and columns implicit indexing for selection by position.
- **[   :   ,   :   ]**: Access rows by integer or labels, or access columns by labels (columns can be omitted).

```python
data = pd.DataFrame(np.arange(20).reshape((4, 5)),
                    index=['Ohio', 'Colorado', 'Utah', 'New York'],
                    columns=['zero', 'one', 'two', 'three', 'four'])
data
```

```python
data.loc['Ohio':'Colorado','two':'four']
```

```python
data[1:] # equals to
# data.loc['Colorado':]
# data.iloc[1:]
```

```python
data[0:2] #--> [] by integer
# equals to
# data.loc['Ohio':'Colorado']
# data.iloc[0:2]
# data['Ohio':'Colorado'] # --> [] by label
```

```python
data[['one']]
```

```python
data[['one','two']]
```

```python
# data.loc['Colorado':'Utah','two':]# equal to
data.iloc[1:3,2:]
```

```python
data.loc[:,'one'] # equal to data['one']
```

```python
data[['three', 'two', 'one']] # equals to
# data.loc[:,['three', 'two', 'one']]
# data.iloc[:,[2,1,0]]
```

```python
data.loc[:, 'two':] # or
#data.iloc[:, 2:]
# data['two':] does not work
```

```python
data.loc['Colorado']
# data['Colorado'] # does not work >>> None of [Index(['Colorado'], dtype='object')] are in the [columns] <<< --> in [] single labels are only for columns
# data.iloc[1]
```

```python
data.loc[['Colorado']]
# data['Colorado':'Colorado']
# data.iloc[[1]]
```

```python
# Row and columns selection
# Examples
data.loc['Ohio':'Colorado', 'three'] # data.loc[[row], [col]] -> label
data.loc['Colorado', ['two', 'three']] # data
data.loc['Ohio':'Colorado', 'three':'four']
data.iloc[[1, 2], 1:2]
data.loc[:'Utah', 'two']
data.iloc[0:2, :] # data.iloc[[row], [col]] -> posizionale
data.iloc[2:3, [3, 0, 1]] # Utah, cols: four, one, two
```

Why use loc? Differences between [] and .loc[]

1.   using **.loc**, the execution time is more than 3X times faster!

2.   loc is explicit: df.loc[row_indexer, column_index] can select rows and columns.

 df[indexer] can only select rows or columns depending on the type of values in indexer and the type of column values the dataframe has.

3.  When a slice is passed to **df.loc** the **end-points are included** in the range. When a slice is passed to **df[...]**, the slice is interpreted as a **half-open** interval.

```python
x = pd.DataFrame(np.arange(9).reshape(3,3))
x
```

```python
x.loc[1:2].index.values
```

```python
x[1:2].index.values
```

### Function Application and Mapping

- *apply*: Apply a function along an axis of the DataFrame
- *map*: Apply a function to a Dataframe elementwise. This method applies a function that accepts and returns a scalar to every element of a DataFrame

```python
frame = pd.DataFrame(np.random.randn(4, 3), columns=list('bde'),
                     index=['Utah', 'Ohio', 'Texas', 'Oregon'])
frame
```

```python
frame['b'].max()
```

```python
frame['b'].min()
```

```python
frame
```

```python
def diff_max_min(x): # x pd.Series
  return x.max() - x.min()
```

```python
frame.apply(diff_max_min)
```

```python
frame.apply(lambda x: x.max() - x.min())
```

```python
frame.apply(diff_max_min, axis='columns')
```

```python
def get_power_2(x): # x valore
  return np.power(x, 2)

frame.map(get_power_2)
```
