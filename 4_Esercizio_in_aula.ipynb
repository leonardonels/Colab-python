1. Inserisci in un dataframe il contenuto del file "dipendenti.csv".

2. Crea una nuova colonna chiamata **Livello Esperienza**, utilizzando la colonna **Anni di esperienza**. Se l'esperienza è minore di 5 anni, mappa il valore a "Junior", tra 5 e 10 a "Mid" e più di 10 a "Senior".

```python
bins=[0,5,10,100]
exp_lvl=["Junior", "Mid", "Senior"]
pd.cut(df["Anni di esperienza"], bins, labels=exp_lvl)
```

```python
def find_level(anni):
  if anni < 5:
    return 'Junior'
  elif anni >= 10:
    return 'Senior'
  else:
    return 'Mid'
```

```python
df['Livello di esperienza']=df['Anni di esperienza'].map(find_level)
df
```

3. Sostituisci i nomi dei dipartimenti con le seguenti abbreviazioni: Vendite -> 'VEN', Information Technology -> 'IT', Human Resources -> 'HR'.

```python
def label_short(dipartimento):
  if dipartimento == 'Vendite':
    return 'VEN'
  elif dipartimento == 'Information Technology':
    return 'IT'
  else:
    return 'HR'
```

```python
df['Dipartimento']=df['Dipartimento'].map(label_short)
df
```
