```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mathchi/churn-for-bank-customers")

print("Path to dataset files:", path)

df = pd.read_csv(f'{path}/churn.csv')
```

```python
df
```

```python
df['Exited'].value_counts()
```

```python
df.isnull().sum().sum()
```

```python
df['Gender']=df['Gender'].map({'Male':0,'Female':1})
df
```

```python
enc = preprocessing.OneHotEncoder()
enc.fit(np.array(df['Geography']).reshape(-1,1))
```

```python
enc.transform(np.array(df['Geography']).reshape(-1,1)).toarray()
```

```python
enc.get_feature_names_out()
```

```python
enc.categories_
```

```python
new_features = []
for i in enc.categories_:
  for j in i:
    new_features.append(j)


new_df = enc.transform(np.array(df['Geography']).reshape(-1,1)).toarray()
new_df_final=pd.DataFrame(new_df, columns=new_features)
new_df_final
```

```python
for i in enc.categories_:
  for j in i:
    df[f'{j}']=new_df_final[f'{j}']

df
```

```python
df.drop(['RowNumber', 'CustomerId', 'Surname', 'Geography'], axis=1, inplace=True)
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('Exited', axis=1), df['Exited'], test_size=0.25, random_state=0)
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
accuracy_score(y_test, dtc.predict(X_test))
```

```python
df['Loyalty']=(df['EstimatedSalary']*df['Tenure'])/2
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('Exited', axis=1), df['Exited'], test_size=0.25, random_state=0)
dtc=DecisionTreeClassifier()
dtc.fit(X_train, y_train)
accuracy_score(y_test, dtc.predict(X_test))
```

```python
df['Investement']=(df['Balance']*df['NumOfProducts'])
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('Exited', axis=1), df['Exited'], test_size=0.25, random_state=0)
dtc=DecisionTreeClassifier()
dtc.fit(X_train, y_train)
accuracy_score(y_test, dtc.predict(X_test))
```
